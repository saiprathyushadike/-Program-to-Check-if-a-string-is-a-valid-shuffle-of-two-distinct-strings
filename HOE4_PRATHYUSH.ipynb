EXAPLE 4.13

```python
x = np.arange(0,10,0.2)
plt.subplot(2,2,1)
plt.plot(x,np.sin(x), 'o--r', x, np.cos(x), 'x-b')

myfont = {'family':'serif',
          'color':'darkred',
          'size':16}
plt.xlabel('x-axis',fontdict=myfont)
plt.ylabel('y-axis',fontdict=myfont)
plt.title('Sine and Cosine Waves',fontdict=myfont , loc='left')
plt.legend(['Sine','Cosine'])

plt.grid(axis = 'y', color = 'r', linestyle = '--', linewidth = 0.5)
plt.xlim(0,10)
plt.xticks(np.arange(0,11,1))

y = np.array([23,45,56,12])
plt.subplot(2,2,2)
lable = ["apples","Bananas","Blueberries","Cherry"]
myexplode = [0.2,0,0,0]
clrs = ["g","y","r","b"]
plt.pie(y,labels = lable, explode = myexplode, colors = clrs, shadow = True)
plt.title('Fruits combo for diet')
plt.legend(title='Fruits', loc = 'upper left')

catg = ['A','B','C', 'D','E']
plt.subplot(2,2,3)
Val = [23,45,56,12,39]
plt.bar(catg,Val,color = 'skyblue', edgecolor='k',width=1)

X = np.random.standard_normal(1000)
plt.subplot(2,2,4)
plt.hist(X, bins=30, color='yellow', edgecolor='black')
plt.title('30 bins')


plt.show()

```

EXAMPLE 4.14 & 15

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 15.5], marker='o')
plt.title("Example of Text and Annotation")
plt.text(2, 17, 'Simple text to label the curve',
         fontdict={'fontsize': 12, 'color': 'blue'})
plt.annotate('Peak Point',
            xy=(3, 25),
            xytext=(2.5, 20),
            arrowprops=dict(facecolor='black',
                          arrowstyle='->'),
            fontsize=12,
            family='serif',
            color='red')

plt.grid(True)

plt.savefig('example_4_14.png', dpi=300, bbox_inches='tight')

plt.savefig('example_4_14.pdf', bbox_inches='tight')

plt.show()

```
